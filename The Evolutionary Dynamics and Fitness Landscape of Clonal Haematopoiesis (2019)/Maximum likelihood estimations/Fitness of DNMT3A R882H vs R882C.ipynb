```python
def error_bars(hist, normed_value, widths):
    
    errors={}
    n=0
    for i in list(hist):
        normalised_hist = i/(normed_value*widths)
        log_hist = np.log(normalised_hist)
        sqrt_hist = math.sqrt(i)
        if sqrt_hist == 1:
            upper_error = 1
            lower_error = 0.9
        if sqrt_hist !=1:
            upper_error = sqrt_hist
            lower_error = sqrt_hist
        normalised_upper_error = upper_error/(normed_value*widths)
        normalised_lower_error = lower_error/(normed_value*widths)
        errors[n]=(normalised_hist[0], normalised_upper_error[0], normalised_lower_error[0])
        n = n+1

    errors_corrected ={}
    for k, v in errors.items():
        binheight = v[0]
        log_binheight = np.log(v[0])
        upper_error = v[1]
        lower_error = v[2]
        log_upper_error = (np.log(upper_error+binheight))-log_binheight
        log_lower_error = log_binheight-(np.log(binheight-lower_error))
        errors_corrected[k] = (log_binheight, log_upper_error, log_lower_error)

    lower_err=[]
    upper_err=[]
    for k, v in errors_corrected.items():
        lower_error = v[2]
        upper_error = v[1]
        lower_err.append(lower_error)
        upper_err.append(upper_error)

    err = [tuple(lower_err),tuple(upper_err)]
    
    return err
```

The purpose of each of these functions:
- **data_for_histogram_** (e.g. data_for_histogram_DNMT3A)
    - this function requires the variant ID (in the format (amino acid (e.g. 617), variant (e.g. V617F))
    - it then imports a .csv file ('all_studies_trimmed_all_genes.csv') which contains all the missense and nonsense variants for all of the studies (trimmed).
    - if the variant is found in this .csv file it stores it in a dictionary ('variant_results') where the key is the row_count in the .csv file and the values are: variant ID, VAF, mutation rate for the variant, age of individual with the variant and the study it was seen in.
<br>
<br>
- **logVAF_list_for_hist_** (e.g. logVAF_list_for_hist_DNMT3A)
    - this function requires the output of the data_for_histogram function as the input (i.e. the dictionary that contains info on the variant of interest.
    - the function looks at the dictionary and separates out the variants from each study in to individual lists for each study (e.g. Jaiswal_VAFs) and records the VAFs (either log or linear depending on the function) for the variant of interest in the list.
<br>
<br>
- **VAF_densities_** (e.g. VAF_densities_DNMT3A)
    - this requires the list of VAFs (for the variant of interest) for the study of interest to be inputted, as well as the number of individuals in the study and how many 'bins' the histogram should have ('Doane method') is used.
    - the function generates a list that contains tuples of (x-axis values for the bin centres for the histogram, the log densities for the VAFs in that bin, the lower and upper error bars).
        - the VAF densities in the histogram are normalised by the number of individuals in the study (study_total) and by 2 x mutation rate of the variant.
<br>
<br>
- **log_hist_data_** (e.g. log_hist_data_DNMT3A)
    - this function draws the histogram for the variant of interest for the study of interest
    - this function requires the mutation-rate increase (calculated from MLE) to be inputted
    - the VAF densities are normalised by the study size as well as 2 x mutation rate x mutation rate increase

# Calculating and plotting the 95% confidence intervals

# Parameter estimation using MLE

```python
#for log(VAF)
def logProbtheory_ages_mut_inc(l, params): #= predicted density (i.e. normalised by 2 x mu)
    "Natural log of the probability of observing a variant within a specific binwidth if able to sequence perfectly"
    total_density=0.0
    N = 9.39358404e+04 #N inferred from DNMT3A R882H (see notebook 'parameter estimation using MLE (DNMT3A R882H etc)')
    sigma = 1.14431010e+01 #sd of ages inferred from DNMT3A R882H (see notebook 'parameter estimation using MLE (DNMT3A R882H etc)')
    s=params[0]
    mut_inc = params[1]
    mean=55 #mean of ages across all studies
    dt=0.1

    total_density=integrate.quad(lambda t: (N/(1-2*np.exp(l))*np.exp(-((np.exp(l))/(((np.exp(s*t)-1)/(2*N*s))*(1-2*np.exp(l))))))*\
                                 ((1/((2*np.pi*(sigma)**2)**0.5))*(np.exp(-(t-mean)**2/(2*(sigma**2))))), mean-2*sigma, mean+2*sigma)
    return np.log(total_density[0])
        
def logProbDataGivenModel_ages_mut_inc(params, data): 
    "This returns the natural log likelihood of the entire data, in specified binwidths, for a given theta and phi"
    total_square_distance = 0
    for datapoint in data:
        if np.exp(datapoint[0])<0.5:
            logfreq = datapoint[0]
            predicted_log_density = logProbtheory_ages_mut_inc(logfreq, params)
            actual_density = np.exp(datapoint[1])
            actual_density_mut_inc = actual_density/params[1]
            actual_log_density_mut_inc = np.log(actual_density_mut_inc)
            square_distance = ((actual_log_density_mut_inc - predicted_log_density)**2)
            total_square_distance = total_square_distance + square_distance

    return total_square_distance   
```

```python
#for linear(VAF)
def logProbtheory_ages_linear_mut_inc(f, params): #= predicted density (i.e. normalised by 2 x mu)
    "Natural log of the probability of observing a variant within a specific binwidth if able to sequence perfectly"
    total_density=0.0
    N = 9.39358404e+04
    sigma = 1.14431010e+01
    s=params[0]
    mut_inc = params[1]
    mean=55
    dt=0.1

    total_density=integrate.quad(lambda t: ((N/(f*(1-(2*f)))*np.exp(-(f)/(((np.exp(s*t))/(2*N*s))*(1-(2*f)))))*\
                                            ((1/((2*np.pi*(sigma)**2)**0.5))*(np.exp(-(t-mean)**2/(2*(sigma**2)))))),mean-2*sigma, mean+2*sigma)
    
    return np.log(total_density[0])

def logProbDataGivenModel_ages_linear_mut_inc(params, data): 
    "This returns the natural log likelihood of the entire data, in specified binwidths, for a given theta and phi"
    total_square_distance = 0
    for datapoint in data:
        if datapoint[0]<0.5:
            freq = datapoint[0]
            predicted_log_density = logProbtheory_ages_linear_mut_inc(freq, params)
            actual_density = np.exp(datapoint[1])
            actual_density_mut_inc = actual_density/params[1]
            actual_log_density_mut_inc = np.log(actual_density_mut_inc)
            square_distance = ((actual_log_density_mut_inc - predicted_log_density)**2)
            total_square_distance = total_square_distance + square_distance

    return total_square_distance   
```
