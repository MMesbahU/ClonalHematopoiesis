```python
# Amino Acid Code:

Amino_acid_abbreviations = {'Ala': 'A', 'Arg': 'R', 'Asn': 'N', 'Asp': 'D',\
                           'Cys': 'C', 'Glu': 'E', 'Gln': 'Q', 'Gly': 'G',\
                           'His': 'H', 'Ile': 'I', 'Leu': 'L', 'Lys': 'K',\
                           'Met': 'M', 'Phe': 'F', 'Pro': 'P', 'Ser': 'S',\
                           'Thr': 'T', 'Trp': 'W', 'Tyr': 'Y', 'Val': 'V', '*': '*'}

Genetic_Code = {'TTT': 'Phe', 'TTC': 'Phe', 'TTA': 'Leu', 'TTG': 'Leu',\
               'TCT': 'Ser', 'TCC': 'Ser', 'TCA': 'Ser', 'TCG': 'Ser',\
               'TAT': 'Tyr', 'TAC': 'Tyr', 'TAA': '*', 'TAG': '*',\
               'TGT': 'Cys', 'TGC': 'Cys', 'TGA': '*', 'TGG': 'Trp',\
               'CTT': 'Leu', 'CTC': 'Leu', 'CTA': 'Leu', 'CTG': 'Leu',\
               'CCT': 'Pro', 'CCC': 'Pro', 'CCA': 'Pro', 'CCG': 'Pro',\
               'CAT': 'His', 'CAC': 'His', 'CAA': 'Gln', 'CAG': 'Gln',\
               'CGT': 'Arg', 'CGC': 'Arg', 'CGA': 'Arg', 'CGG': 'Arg',\
               'ATT': 'Ile', 'ATC': 'Ile', 'ATA': 'Ile', 'ATG': 'Met',\
               'ACT': 'Thr', 'ACC': 'Thr', 'ACA': 'Thr', 'ACG': 'Thr',\
               'AAT': 'Asn', 'AAC': 'Asn', 'AAA': 'Lys', 'AAG': 'Lys',\
               'AGT': 'Ser', 'AGC': 'Ser', 'AGA': 'Arg', 'AGG': 'Arg',\
               'GTT': 'Val', 'GTC': 'Val', 'GTA': 'Val', 'GTG': 'Val',\
               'GCT': 'Ala', 'GCC': 'Ala', 'GCA': 'Ala', 'GCG': 'Ala',\
               'GAT': 'Asp', 'GAC': 'Asp', 'GAA': 'Glu', 'GAG': 'Glu',\
               'GGT': 'Gly', 'GGC': 'Gly', 'GGA': 'Gly', 'GGG': 'Gly'}

base_alternative = {'A': ['T', 'C', 'G'], 'T': ['A', 'C', 'G'], 'G':['A', 'C', 'T'], 'C':['A', 'G', 'T']}
```

```python
Possible_AA_Changes = {}
Possible_Base_Changes = {}
Possible_Codon_Changes = {}
for k, v in Genetic_Code.items():
    codon = k
    base_1 = str(codon[0])
    base_2 = str(codon[1])
    base_3 = str(codon[2])
    new_codons = []
    amino_acid_changes = []
    base_changes = []
    if base_1 in base_alternative.keys():
        alternate_base = base_alternative[base_1]
        for i in alternate_base:
            base_changes.append(i)
            new_codon = i+base_2+base_3
            new_codons.append(new_codon)
    if base_2 in base_alternative.keys():
        alternate_base = base_alternative[base_2]
        for i in alternate_base:
            base_changes.append(i)
            new_codon = base_1+i+base_3
            new_codons.append(new_codon)
    if base_3 in base_alternative.keys():
        alternate_base = base_alternative[base_3]
        for i in alternate_base:
            base_changes.append(i)
            new_codon = base_1+base_2+i
            new_codons.append(new_codon)
    Possible_Codon_Changes[k]=(new_codons[0],new_codons[1], new_codons[2], new_codons[3], new_codons[4],\
                               new_codons[5], new_codons[6], new_codons[7], new_codons[8]) #i.e. {'TTT': ['ATT', 'CTT', 'GTT'...etc]}
    Possible_Base_Changes[k]=(base_changes[0], base_changes[1], base_changes[2], base_changes[3], base_changes[4],\
                              base_changes[5], base_changes[6], base_changes[7], base_changes[8]) #i.e. {'TTT': ['A', 'C', 'G'... etc]}
    
Possible_Amino_Acid_Changes = {}
for k, v in Possible_Codon_Changes.items():
    codon = k
    codon_changes = v
    amino_acid_changes = []
    for i in codon_changes:
        amino_acid_change = Genetic_Code[i]
        amino_acid_changes.append(amino_acid_change)
    Possible_Amino_Acid_Changes[k]=(amino_acid_changes[0], amino_acid_changes[1], amino_acid_changes[2],\
                                    amino_acid_changes[3], amino_acid_changes[4], amino_acid_changes[5],\
                                    amino_acid_changes[6], amino_acid_changes[7], amino_acid_changes[8]) #i.e. {'TTT': ['Ile', 'Leu', 'Val'...etc]}
    
Possible_Amino_Acid_Changes_letters = {}
for k, v in Possible_Amino_Acid_Changes.items():
    codon = k
    codon_changes = v
    amino_acid_changes_letters = []
    for i in codon_changes:
        amino_acid_change_letter = Amino_acid_abbreviations[i]
        amino_acid_changes_letters.append(amino_acid_change_letter)
    Possible_Amino_Acid_Changes_letters[k]=(amino_acid_changes_letters[0], amino_acid_changes_letters[1],\
                                            amino_acid_changes_letters[2], amino_acid_changes_letters[3],\
                                            amino_acid_changes_letters[4], amino_acid_changes_letters[5],\
                                            amino_acid_changes_letters[6], amino_acid_changes_letters[7],\
                                            amino_acid_changes_letters[8]) #i.e. {'TTT': ['I', 'L', 'V'...etc]}
```

```python
#Import mutation rates
import csv
filename="Data files/Trinucleotide_context_mutation_rates.csv" #csv file containing trinucleotide-context specific mutation rates (calculated from Lee-Six et al 2018 data)

with open(filename, 'r') as csvfile:
    read_reader = csv.reader(csvfile)  #csv.reader returns a reader object which will iterate over lines in the csvfile
    row_count=0
    Mutation_rates = {} #create a dictionary where key = trincucleotide site, value = mutation rate

    for row in read_reader:
        if row_count>0:
            site = row[0] #trinucleotide site
            comp_site = row[2] #complementary site
            if site != '':
                mutation_rate = float(row[7])
                Mutation_rates[site] = mutation_rate
                Mutation_rates[comp_site] = mutation_rate
        row_count = row_count+1
```

```python
def mutation_rates_missense_range_of_variants_TP53(lower_range, upper_range):

    #The mutation rate at a base is calculated based on its trinucleotide context.  To obtain the trinucleotide context, the base either side of the mutated base needs to be ascertained.
    #The coding sequence (+ an extra base at each end) of each gene is used when determining the trinucleotide context of the base change.  The problem with using the coding sequence is that
    #if a base change falls at the end of an exon, its neighbouring base will actuall be an intronic base, which may be different to the 'neighbouring' base in the coding sequence.
    #To account for this, there is an "intron_exon_check" function that is called at the intron/exon boundary if the neighbouring base in the intron is different to the neighbouring base in the coding sequence.
    
    intron_exon_boundary_caution_base_TP53 = [74, 75, 96, 375, 672, 782, 919, 993, 1100]
   
    def intron_exon_check(position_in_gene_number):
        if position_in_gene_number == 74:
            gene_pos = 74
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 75:
            gene_pos = 75
            first_base = 'G'
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = TP53_exons_extra_bases[gene_pos+1]
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 96:
            gene_pos = 96
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 375:
            gene_pos = 375
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 672:
            gene_pos = 672
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 782:
            gene_pos = 782
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 919:
            gene_pos = 919
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 993:
            gene_pos = 993
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        if position_in_gene_number == 1100:
            gene_pos = 1100
            first_base = TP53_exons_extra_bases[gene_pos-1]
            middle_base = TP53_exons_extra_bases[gene_pos]
            third_base = 'G'
            triplet = first_base+middle_base+third_base
#             print(triplet)

        return triplet
    
    #TP53 transcript:
    #Downloaded from http://grch37.ensembl.org
    TP53_exons = 'ATGGAGGAGCCGCAGTCAGATCCTAGCGTCGAGCCCCCTCTGAGTCAGGAAACATTTTCAGACCTATGGAAACTACTTCCTGAAAACAACGTTCTGTCCCCCTTGCCGTCCCAAGCAATGGATGATTTGATGCTGTCCCCGGACGATATTGAACAATGGTTCACTGAAGACCCAGGTCCAGATGAAGCTCCCAGAATGCCAGAGGCTGCTCCCCCCGTGGCCCCTGCACCAGCAGCTCCTACACCGGCGGCCCCTGCACCAGCCCCCTCCTGGCCCCTGTCATCTTCTGTCCCTTCCCAGAAAACCTACCAGGGCAGCTACGGTTTCCGTCTGGGCTTCTTGCATTCTGGGACAGCCAAGTCTGTGACTTGCACGTACTCCCCTGCCCTCAACAAGATGTTTTGCCAACTGGCCAAGACCTGCCCTGTGCAGCTGTGGGTTGATTCCACACCCCCGCCCGGCACCCGCGTCCGCGCCATGGCCATCTACAAGCAGTCACAGCACATGACGGAGGTTGTGAGGCGCTGCCCCCACCATGAGCGCTGCTCAGATAGCGATGGTCTGGCCCCTCCTCAGCATCTTATCCGAGTGGAAGGAAATTTGCGTGTGGAGTATTTGGATGACAGAAACACTTTTCGACATAGTGTGGTGGTGCCCTATGAGCCGCCTGAGGTTGGCTCTGACTGTACCACCATCCACTACAACTACATGTGTAACAGTTCCTGCATGGGCGGCATGAACCGGAGGCCCATCCTCACCATCATCACACTGGAAGACTCCAGTGGTAATCTACTGGGACGGAACAGCTTTGAGGTGCGTGTTTGTGCCTGTCCTGGGAGAGACCGGCGCACAGAGGAAGAGAATCTCCGCAAGAAAGGGGAGCCTCACCACGAGCTGCCCCCAGGGAGCACTAAGCGAGCACTGCCCAACAACACCAGCTCCTCTCCCCAGCCAAAGAAGAAACCACTGGATGGAGAATATTTCACCCTTCAGATCCGTGGGCGTGAGCGCTTCGAGATGTTCCGAGAGCTGAATGAGGCCTTGGAACTCAAGGATGCCCAGGCTGGGAAGGAGCCAGGGGGGAGCAGGGCTCACTCCAGCCACCTGAAGTCCAAAAAGGGTCAGTCTACCTCCCGCCATAAAAAACTCATGTTCAAGACAGAAGGGCCTGACTCAGACTGA'
    
    #Split the sequence into codons 
    TP53_codons = [TP53_exons[i:i+3] for i in range(0, len(TP53_exons), 3)]

    #TP53 transcript:
    #Downloaded from http://grch37.ensembl.org
    TP53_exons_extra_bases = 'CATGGAGGAGCCGCAGTCAGATCCTAGCGTCGAGCCCCCTCTGAGTCAGGAAACATTTTCAGACCTATGGAAACTACTTCCTGAAAACAACGTTCTGTCCCCCTTGCCGTCCCAAGCAATGGATGATTTGATGCTGTCCCCGGACGATATTGAACAATGGTTCACTGAAGACCCAGGTCCAGATGAAGCTCCCAGAATGCCAGAGGCTGCTCCCCCCGTGGCCCCTGCACCAGCAGCTCCTACACCGGCGGCCCCTGCACCAGCCCCCTCCTGGCCCCTGTCATCTTCTGTCCCTTCCCAGAAAACCTACCAGGGCAGCTACGGTTTCCGTCTGGGCTTCTTGCATTCTGGGACAGCCAAGTCTGTGACTTGCACGTACTCCCCTGCCCTCAACAAGATGTTTTGCCAACTGGCCAAGACCTGCCCTGTGCAGCTGTGGGTTGATTCCACACCCCCGCCCGGCACCCGCGTCCGCGCCATGGCCATCTACAAGCAGTCACAGCACATGACGGAGGTTGTGAGGCGCTGCCCCCACCATGAGCGCTGCTCAGATAGCGATGGTCTGGCCCCTCCTCAGCATCTTATCCGAGTGGAAGGAAATTTGCGTGTGGAGTATTTGGATGACAGAAACACTTTTCGACATAGTGTGGTGGTGCCCTATGAGCCGCCTGAGGTTGGCTCTGACTGTACCACCATCCACTACAACTACATGTGTAACAGTTCCTGCATGGGCGGCATGAACCGGAGGCCCATCCTCACCATCATCACACTGGAAGACTCCAGTGGTAATCTACTGGGACGGAACAGCTTTGAGGTGCGTGTTTGTGCCTGTCCTGGGAGAGACCGGCGCACAGAGGAAGAGAATCTCCGCAAGAAAGGGGAGCCTCACCACGAGCTGCCCCCAGGGAGCACTAAGCGAGCACTGCCCAACAACACCAGCTCCTCTCCCCAGCCAAAGAAGAAACCACTGGATGGAGAATATTTCACCCTTCAGATCCGTGGGCGTGAGCGCTTCGAGATGTTCCGAGAGCTGAATGAGGCCTTGGAACTCAAGGATGCCCAGGCTGGGAAGGAGCCAGGGGGGAGCAGGGCTCACTCCAGCCACCTGAAGTCCAAAAAGGGTCAGTCTACCTCCCGCCATAAAAAACTCATGTTCAAGACAGAAGGGCCTGACTCAGACTGAC'
    
    #List of all amino acid positions (across whole coding region) covered (amino acid position minus 1 (as python starts at 0))
    TP53_missense_codons_DICT={}
    for i in range(lower_range, upper_range):
        TP53_missense_codons_DICT[i+1]= (TP53_codons[i])

    calls_TP53_missense = {}
    for k,v in TP53_missense_codons_DICT.items(): #iterating over the list of codon numbers (and codon sequences)
        if v in Possible_Amino_Acid_Changes.keys(): #assign the list of all the possible amino acid changes to each codon they called
            changes = Possible_Amino_Acid_Changes_letters[v]
    #         print(Possible_Amino_Acid_Changes.values())
            calls_TP53_missense[k]=(v, changes)

    #Changing the original amino acid to a Word (e.g. Phe) rather than a codon
    calls_TP53_missense_extra = {}

    for k, v in calls_TP53_missense.items():
        if v[0] in Genetic_Code:
            Amino_acid_change = Genetic_Code.get(v[0])
            calls_TP53_missense_extra[k] = (Amino_acid_change, v[0], v[1])
            #e.g. 12: (Phe, 'K', '*', 'T' etc..)

    #Changing the original amino acid to a LETTER (rather than the codon)
    calls_TP53_missense_extra2 = {}

    for k, v in calls_TP53_missense_extra.items():
        if v[0] in Amino_acid_abbreviations:
            Amino_acid_abbrev = Amino_acid_abbreviations.get(v[0])
            calls_TP53_missense_extra2[k] = (Amino_acid_abbrev, v[1], v[2])
            #e.g. 12: ((F), ('K', '*', 'T' etc..))

    #Create a dictionary that contains details of the positions of missense variants at that codon
    TP53_position_in_list_dict_missense = {}

    for k, v in calls_TP53_missense_extra2.items():
        original_AA = v[0]
        possible_AA_changes = v[2]
        possible_AA_changes_string = ''.join(v[2])
        positions_in_list = []
        for i in possible_AA_changes:
            if i != original_AA: #not synonymous
                if i != '*': #not missense
                    position_in_list=-1
                    missense_change = i
                    new_aa_list = []
                    while True:
                        position_in_list = possible_AA_changes_string.find(missense_change, position_in_list+1)
        #               print(position_in_list)
                        if position_in_list == -1: break
                        positions_in_list.append(position_in_list)
                        positions_in_list = set(positions_in_list) #convert into a 'set' to remove duplicate positions
                        positions_in_list = list(positions_in_list) #convert back to a list so can be added to dictionary
                    for pos in positions_in_list:
                        new_aa = possible_AA_changes_string[pos]
                        new_aa_list.append(new_aa)
                    TP53_position_in_list_dict_missense[k]=(positions_in_list, v, new_aa_list)

        aa_positions_variants_called = []
        for k, v in TP53_position_in_list_dict_missense.items():
            aa_positions_variants_called.append(k)
    # print('positions at which missense variants called = ', aa_positions_variants_called)

    #Create a dictionary that contains the position in the list of possible AA where the variant is and what it changes to
    TP53_bases_missense = {}
    for k, v in TP53_position_in_list_dict_missense.items():
        original_a_a = v[1][0]
        codon_seq = v[1][1]
        base_changes=[]
        new_a_a_list = v[2]
        if codon_seq in Possible_Base_Changes.keys():
            list_of_possible_bases = Possible_Base_Changes.get(codon_seq)
        for i in v[0]:
            base_change_position = i
            base_change = list_of_possible_bases[base_change_position]
            base_changes.append(base_change)
        TP53_bases_missense[k]=(v, base_changes, original_a_a, new_a_a_list)

    #Create dictionary which contains original codon seq and details on which base is changed
    TP53_bases_missense2 = {}

    for k, v in TP53_bases_missense.items():
        base_change_positions = v[0][0]
        codon_positions = []
        positions_in_gene = []
        new_a_a_list = v[3]
        for i in base_change_positions:
            base_change_position = i
            if base_change_position <= 2:
                codon_pos = 1
            if 2 < base_change_position <= 5:
                codon_pos = 2
            if 5 < base_change_position <= 8:
                codon_pos = 3
            codon_positions.append(codon_pos)
        TP53_bases_missense2[k] = (v, codon_positions)

    #Create dictionary which contains original codon seq, details on which base is changed and it's position in the coding region
    TP53_bases_missense3 = {}

    for k, v in TP53_bases_missense2.items():
        codon_positions = v[1]
        base_changes = v[0][1]
        original_a_a = v[0][2]
        new_aa_list = v[0][0][2]
        original_bases = []
        positions_in_gene = []
        codon_number = k
        for i in codon_positions:
            codon_pos = i
            position_in_gene = ((int(k)-1)*3)+codon_pos #codon number minus 1 (e.g. codon 307 -1 = 306, then times by 3 takes you to beginning of codon 307 (then add the position))  
            positions_in_gene.append(position_in_gene)
            original_base = (v[0][0][1][1][codon_pos-1])
            original_bases.append(original_base)
        TP53_bases_missense3[k] = (v[0][0][1], positions_in_gene, original_bases, base_changes, codon_number, original_a_a, new_aa_list)

    # Create dictionary which contains original codon seq, details on which base is changed and its neighbouring bases
    TP53_bases_missense4 = {}

    for k, v in TP53_bases_missense3.items():
        position_in_gene = v[1]
        sites_trio_missense=[]
        new_a_a_list = v[6]

        for i in position_in_gene:
    #         print(i)
            if i in intron_exon_boundary_caution_base_TP53:
#                 print('positions at intron_exon_boundary =',i)
                adapted_trio = intron_exon_check(i)
                sites_trio_missense.append([adapted_trio])
                TP53_bases_missense4[k] = (v, sites_trio_missense)

            else:
                for i in position_in_gene:
                    gene_pos = i
                    site_trio_missense = [TP53_exons_extra_bases[gene_pos:gene_pos+3] for gene_pos in range (gene_pos-1, gene_pos, 3)]
                    sites_trio_missense.append(site_trio_missense)
                    TP53_bases_missense4[k] = (v, sites_trio_missense)

    #Create a dictionary containing pos in coding region, original base, new base, base affected with its neighbours
    #e.g. 938, G, A, TGG = base 938 on coding region, original base = G, changes to base A, base with neighbours either side = TGG
    TP53_sites_missense5 = {}

    a=0
    for k, v in TP53_bases_missense4.items():
        original_a_a = v[0][0][0]
        new_a_a = v[0][6]
        codon_n = v[0][1] #base number in codon (out of 3 x 912)
        codon_n_912 = v[0][4]
        original_base = v[0][2]
        new_base = v[0][3]
        site_trio = v[1]
        info=[]

        change = 0
        for i in codon_n:
            codon_number = i
            orig_base = original_base[change]
            new_amino_acid = new_a_a[change]
            n_base = new_base[change]
            site_tri = site_trio[change]
            info.append([codon_number, orig_base, n_base, site_tri, codon_n_912, original_a_a, new_amino_acid])
            change = change+1
        TP53_sites_missense5[a]=info
        a=a+1
    #     print(TP53_sites_missense5)

    # Write in a way that can iterate over mutation rates (i.e. put in to form C[C>A]G etc)   
    TP53_sites_missense6={}

    separating_positions = {} # separate out all the possible changes to just 1 effective item in the list
    a=0
    for k, v in TP53_sites_missense5.items():
        for i in v: #because some contain more than 1 change from that codon:
            separating_positions[a]=i
            a=a+1

    for k, v in separating_positions.items():
        base_pos = v[0]
        original_base = v[1]
        new_base = v[2]
        site_trio = v[3]
        codon_n_912 = v[4]
        original_a_a = v[5]
        new_a_a = v[6]
        sequence = (str(original_base),str(new_base))
        changes = '>'.join(sequence)
        left_base = site_trio[0][0]
        right_base = site_trio[0][2]
        sequence2 = (changes, right_base)
        changes2 = ']'.join(sequence2)
        sequence3 = (left_base, changes2)
        changes3 = '['.join(sequence3)
        TP53_sites_missense6[k] = (v[0],changes3, codon_n_912, original_a_a, new_a_a) #site_trio = the affected base and its neighbours

    #Assign the mutation rates to each site...
    TP53_mutation_rates_missense = {}
    for k, v in TP53_sites_missense6.items():
        base_change_info = v[1]
        if base_change_info in Mutation_rates.keys():
            mutation_rate = Mutation_rates.get(base_change_info)
            original_a_a = v[3]
            new_a_a = v[4]
            aa_positioning = v[2]
            variant_id = str(original_a_a)+str(aa_positioning)+str(new_a_a)
            TP53_mutation_rates_missense[k] = (v[0],v[1], mutation_rate, variant_id)

    TP53_mutation_rate_missense_sum = []

    mutation_rates_list = []
    for k, v in TP53_mutation_rates_missense.items():
        codon_position_in_gene=(v[0])
        variant = v[3]
        site = v[1]
        mutation_rate_value = v[2] 
        mutation_rates_list.append((v[0], variant, site, mutation_rate_value))
        TP53_mutation_rate_missense_sum.append(mutation_rate_value)

    #Get the average of all the mutation rates
    TP53_mutation_rate_missense_sum = []
    for k, v in TP53_mutation_rates_missense.items():
        mut_rate = v[2]
        TP53_mutation_rate_missense_sum.append(mut_rate)

    TP53_Summed_mutation_rate_missense_added = np.sum(TP53_mutation_rate_missense_sum)

    print('Total summed mutation rate across TP53 missense for this study = ', TP53_Summed_mutation_rate_missense_added)

    mutation_rates_list_all = mutation_rates_list
    return TP53_Summed_mutation_rate_missense_added
```

# Actual number of observations of TP53 variants (inside vs outside DNA binding domain)

- Look at each variant seen across the studies and count how many times it was observed.

```python
def TP53_variants_by_study(study_name):

    filename = "Data files/all_studies_trimmed_all_genes.csv"
    with open(filename, 'r') as csvfile:
        read_reader = csv.reader(csvfile)
        row_count = 0
        missense_TP53_DBD = []
        missense_TP53_nonDBD = []
        nonsense_TP53_DBD = []
        nonsense_TP53_nonDBD = []

        for row in read_reader:
#             print(row)
            if row_count > 0:
                VAF = float(row[0])
                age = row[1]
                variant = row[2]
                codon_number = int(re.search(r'\d+', variant).group(0)) #extract the codon number
                gene = row[3]
                study = row[4]
                if study == study_name:
                    if gene == 'TP53':
                        if 100 <= codon_number <= 300:
                            if variant[-1]=='*': #If the final part of the variant is * (i.e. nonsense)
                                nonsense_TP53_DBD.append(variant)
                            else:
                                missense_TP53_DBD.append(variant)
                        else:
                            if variant[-1]=='*': #If the final part of the variant is * (i.e. nonsense)
                                nonsense_TP53_nonDBD.append(variant)
                            else:
                                missense_TP53_nonDBD.append(variant)
            row_count = row_count+1
            
    return missense_TP53_DBD, nonsense_TP53_DBD, missense_TP53_nonDBD, nonsense_TP53_nonDBD
```

# Plotting the relative frequencies of mutations (controlled for mutation rates)

- If the number of missense variants observed in the DBD is Poisson distributed with mean 5 and the number of missense variants outside the DBD is poisson distributed with proportional mean of 5 x (missense non DBD mutation rate/ missense DBD mutation rate) (i.e. null hypothesis is that there is no difference in the number of observations between the DBD and non-DBD), what is the probability that we only observe 1 variant outside the DBD?

```python
def error_bars(observed_variants, mutation_rate):
    
    errors={}
    n=0
    for i, j in zip(observed_variants, mutation_rate):
        normalised_hist = i/j
        sqrt_hist = math.sqrt(i)
        upper_error = sqrt_hist
        lower_error = sqrt_hist
        normalised_upper_error = upper_error/j
        normalised_lower_error = lower_error/j
        errors[n]=(normalised_hist, normalised_upper_error, normalised_lower_error)
        n = n+1
        
    errors_corrected ={}
    for k, v in errors.items():
        binheight = v[0]
        upper_error = v[1]
        lower_error = v[2]
        corrected_upper_error = (upper_error+binheight)-binheight
        corrected_lower_error = binheight-(binheight-lower_error)
        errors_corrected[k] = (binheight, upper_error, lower_error)

    lower_err=[]
    upper_err=[]
    for k, v in errors_corrected.items():
        lower_error = v[2]
        upper_error = v[1]
        lower_err.append(lower_error)
        upper_err.append(upper_error)

    err = [tuple(lower_err),tuple(upper_err)]
    
    return err
```

```python
def plotting_frequencies(study_name, observations, mutation_rates, ax):

    #colours
    missense = c6
    nonsense = c10
    DBD_missense = missense
    nonDBD_missense = missense
    DBD_nonsense = nonsense
    nonDBD_nonsense = nonsense

    axisfont=20
    domainfont = 12
    titlefont=12
    axislabelfont=12

    scale = 1
    m_size = 18*scale
    m_width = 1.5*scale
    c_size = 5
    c_thick = 1.5
    e_width = 1.5
    
    missense_DBD_number = float(observations[0])
    missense_nonDBD_number = float(observations[1])
    nonsense_DBD_number = float(observations[2])
    nonsense_nonDBD_number = float(observations[3])
    
    missense_DBD_mu = float(mutation_rates[0])
    missense_nonDBD_mu = float(mutation_rates[1])
    nonsense_DBD_mu = float(mutation_rates[2])
    nonsense_nonDBD_mu = float(mutation_rates[3])
    
    missense_DBD_freq = missense_DBD_number/missense_DBD_mu
    missense_nonDBD_freq = missense_nonDBD_number/missense_nonDBD_mu
    nonsense_DBD_freq = nonsense_DBD_number/nonsense_DBD_mu
    nonsense_nonDBD_freq = nonsense_nonDBD_number/nonsense_nonDBD_mu

    colors_dict = {'missense_DBD': DBD_missense, 'missense_nonDBD': nonDBD_missense, 'nonsense_DBD': DBD_nonsense, 'nonsense_nonDBD': nonDBD_nonsense}

    x = ['missense_DBD', 'missense_nonDBD', 'nonsense_DBD', 'nonsense_nonDBD']
    y = [missense_DBD_freq, missense_nonDBD_freq, nonsense_DBD_freq, nonsense_nonDBD_freq]

    ax.bar(x, y, color = [DBD_missense, nonDBD_missense, DBD_nonsense, nonDBD_nonsense])

    errors = error_bars([missense_DBD_number, missense_nonDBD_number, nonsense_DBD_number, nonsense_nonDBD_number], \
                        [missense_DBD_mu, missense_nonDBD_mu, nonsense_DBD_mu, nonsense_nonDBD_mu])

    ax.errorbar(x, y, yerr= errors, ecolor = grey3, fmt = 'none', \
             elinewidth = e_width, capsize = c_size, capthick = c_thick, zorder=3)

    #Only show the required axis lines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    for axis in ['bottom','left']:
        ax.spines[axis].set_linewidth(2)

    for axis in ['bottom','left']:
        ax.spines[axis].set_color(grey3)

    ax.set_ylabel('number of observations / $\u03BC$', fontsize = axisfont, labelpad = 10)    
    ax.set_xlabel('Variant type and TP53 domain', fontsize = axisfont, labelpad = 10)
    ax.set_title(study_name, fontsize = axisfont, fontweight = 'bold')

    # #x-axis ticks
    x1_major_ticks = ['missense_DBD', 'missense_nonDBD', 'nonsense_DBD', 'nonsense_nonDBD']
    x1_major_tick_labels = ['Miss DBD', 'Miss non-DBD', 'Nons DBD', 'Nons non-DBD']
    ax.set_xticks(x1_major_ticks)
    ax.set_xticklabels(x1_major_tick_labels, fontsize = axisfont-5)
    
    #labels
    labels = ["n = %d" % i for i in observations]

    rects = ax.patches
    
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height*1.1, label,
                ha='center', va='bottom', fontsize = axisfont-5)
    
    print()
    
    #seeing if difference between missense DBD and missense non-DBD is significant:
    nonDBD_mean_null = missense_DBD_number*(missense_nonDBD_mu/missense_DBD_mu)
    p_value = poisson.cdf(missense_nonDBD_number, nonDBD_mean_null)
    print('Poisson exact test p-value missense DBD vs missense non-DBD '+str(study_name)+'= ', p_value)
    print("")    
    
    return p_value, ax
```
