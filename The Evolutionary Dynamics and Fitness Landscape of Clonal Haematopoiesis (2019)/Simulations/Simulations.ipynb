# Define useful functions for the simulations

+ Each mutation has an ID (ID) and a growth advantage (s) as a tuple
``[(ID1, s1)]``
+ Each clone contains a certain number of mutations, so could be defined as a list of the mutation, fitness effect tuples
``[(ID1, s1),(ID2, s2)....]``
+ Each clone also needs an associated size (n) (e.g. 50 cells) -> ``[[(ID1, s1),(ID2, s2)....], n]``
+ Total population is a list of the different clones (with a different set of mutations within it): ``[[[(ID1, s1),(ID2, s2)....], n], [[(ID1, s1),(ID2, s2)....], n]]]``

```python
# define function 'divide'
def divide(clone_info, dt, B0, D0):  
    #dt = a small interval of time, 
    #D0 = symmetric rate to differentiated
    #B0 = symmetric rate to self renewal
    #B  = modified birth rate including fitness advantage
    
    mutations=clone_info[0] # [(ID1, s1)]
    clone_size=clone_info[1] # n
    fitness_effects=[i[1] for i in mutations]
    F=sum(fitness_effects)
    B = B0*(1.0+F) #B = birth rate, which is 1 + the overall fitness effect of the mutations within a particular clone
    number_of_births = np.random.poisson(clone_size*B*dt) #pulls a random number from the Poisson distribution with/
    #mean of clone_size x birth rate x interval of time
    number_of_deaths = np.random.poisson(clone_size*D0*dt)
    new_clone_size = clone_size + number_of_births - number_of_deaths
    return [mutations, new_clone_size]
```

```python
#define function 'mutation_fitness' of the form:
# sb=0.05 #beneficial fitness effect
# sd=0.00 #deleterious fitness effect
# Pb=2/3 #probability of beneficial fitness effect
# Pn=1/3 #probability of neutral fitness effect
# Pd=0.00 #probability of deleterious fitness effect
# DFE = [[sd, Pd], [0.0, Pn], [sb, Pb]] #distribution of fitness effects doesn't need to be normalized (because normalized in function below)

def mutation_fitness(DFE):
    total_prob=sum([i[1] for i in DFE])
    normalized_DFE=[[i[0], i[1]/total_prob] for i in DFE]
    
    random_number=np.random.random()
    cumulative = 0.0
    for [s,P] in normalized_DFE:
        cumulative = cumulative+P
        if random_number<cumulative:
            fitness_effect=s
            break
    return fitness_effect
```

```python
#define function 'mutate'
def mutate(clone_info, dt, u, last_mutation_ID, DFE): #u=mutation rate
    mutations=clone_info[0] #(ID, s)
    clone_size=clone_info[1]
    number_of_mutations = np.random.poisson(clone_size*u*dt)
    list_of_new_clones=[]
    for i in range(number_of_mutations):
        last_mutation_ID = last_mutation_ID+1
        new_fitness_effect = mutation_fitness(DFE)
        new_clone = [mutations+[(last_mutation_ID, new_fitness_effect)],1]
        list_of_new_clones.append(new_clone)
    return (list_of_new_clones, last_mutation_ID)
```

```python
#define function 'pop_size'
def pop_size(pop):
    total=0
    for clone in pop:
        sub_pop_size=clone[-1]
        total=total+sub_pop_size
    return total        
```

```python
#define function 'sequence' (to include R882 variants)
def sequence(pop, mean_depth):
    N=pop_size(pop)
    variant_dict={}
    list_of_variants=[]
    
    #first determine total freq of variant across all clones and record clones in which it arises
    for (clone1, size) in pop:
        if len(clone1)>1: #ensures non-WT clones only
            clone=copy.deepcopy(clone1)
            clone.pop(0)
            
            for mutation in clone:
                
                #update mutations frequency
                mID=mutation[0]
                if mID in variant_dict:
                    true_freq=variant_dict[mID]['true_freq']+0.5*(size/N)
                    variant_dict[mID]['true_freq']=true_freq
                else:
                    variant_dict[mID]={'true_freq':0.5*(size/N)}
                    #mutation type    
                    if 0.00<mutation[1]<R882:
                        mutation_type="n" # non-synonymous (with a range of s)
                    elif mutation[1]==0.0:
                        mutation_type="s" #synonymous
                    elif mutation[1]==R882:
                        mutation_type="R882" #R882
                    variant_dict[mID]['mutation_type']=mutation_type
                    
    for m, e in variant_dict.items():
        
        mutation_type=e['mutation_type']
        true_freq=e['true_freq']
        

        depth1 = int(mean_depth*np.exp(np.random.normal(0.0, .3)))
        expected_reads1=true_freq*depth1
        reads1=np.random.poisson(expected_reads1)

        depth2 = int(mean_depth*np.exp(np.random.normal(0.0, .3)))
        expected_reads2=true_freq*depth2
        reads2=np.random.poisson(expected_reads2)

        if reads1<4:
            reads1=0
        if reads2<4:
            reads2=0

        freq1=reads1/depth1
        freq2=reads2/depth2

        list_of_variants.append({"mutation_ID":m, \
                                 "mutation_type":mutation_type, \
                                 "true_freq": true_freq, \
                                 "freq1":freq1,\
                                 "reads1":reads1, \
                                 "depth1": depth1, \
                                 "freq2":freq2,\
                                 "reads2":reads2, \
                                 "depth2": depth2
                                })
    return list_of_variants     
```

# Dynamics of simulation

What are the theoretical expectations?

+ The SFS to use is:$\rho(f) = \frac{\theta}{f(1-2f)} \exp\left(\frac{f/\phi}{1-2f} \right)$
+ where $\theta = N\mu \tau$ where $u$ is the mutation rate
+ where $\phi = (\exp(st)-1)/(2Ns\tau)$ 
+ In plain English this means $\theta = N \times \text{total mutation rate per symmetric cell division}$
+ In plain English the $s \tau$ that enters in the denominator in $\phi$ is the fitness effect per symmetric cell division.
+ In the simulations:
    + below $s_b$ is defined per symmetric cell division i.e. it is $s \tau$
    + $u$ is the mutation rate per cell division
    + $\lambda$ is the HSC division rate per year
    + $B_0 = D_0$ is the fraction of cell divisions that are symmetric self-renewals or symmetric differentiations.
+ Therefore:
    + $\theta = N (u / D_0)$
    + $\phi = (\exp(s_b \times t \times \lambda \times D_0 )-1)/(2Ns)$
    + where $t$ is age in years. 

# Simulation of variant trajectories over lifetime (e.g. Figure 1d)

```python
#EXAMPLE PARAMETERS FOR SIMULATION
eq_population_size = 10**5 #number of stem cells (HSCs)
lifespan = 500 #measured in cell divisions (e.g. 100 = 100 cell divisions)
dt=0.1 # measured in units of the overall cell division rate
u=3*10**(-6) #mutation rate
lam=5 # HSC divisions per year (symmetric and asymmetric)
last_mutation_ID=0
average_depth=3000 #sequencing depth
number_of_individuals=5
mean_age=60
SD_age=0.001

#FITNESS EFFECTS (per year)
sb=0.01 #beneficial fitness effect
sd=0.00 #deleterious fitness effect
R882 = sb*10 #fitness effect of R882 variant

Pb=2/3 #probability of beneficial variant
Pn=0.25 #probability of neutral variant
PR882 = 1-(Pb+Pn) #probability of R882 variant
Pnonsyn = ((1-Pn)-PR882)/6 #probability of nonsynonymous variant
Pd=0.00 #probability of deleterious variant

#exponential distribution of fitness effects for nonsynonymous variants
s_max=0.09 #maximum fitness effect for nonsynonymous variant
s_range=np.linspace(sb, s_max, 9.9) #i.e. nonsynonymous beneficial fitness effects range from sb (=0.01) to s_max (0.099)
d = 0.03
weights=np.array([np.exp(-s/d) for s in s_range])
normalization=np.sum(weights)
norm_weights=(weights/normalization)*Pnonsyn

DFE = [[s, w] for s, w in zip(s_range, norm_weights)]
DFE.append([R882, PR882])
DFE.append([0.0, Pn])
```

```python
#SIMULATION 
population_dict={}

clone_histories={}

timestamp=str(datetime.datetime.now())
datafilename='watson_et_al_D='+"{0:.2e}".format(average_depth)\
+'_N='+"{0:.2e}".format(eq_population_size)\
+'_U='+"{0:.3e}".format(u)\
+'_s='+str(sb)\
+'_Pb='+"{0:.3f}".format(Pb)\
+'exp_s_range0.01_0.09'\
+'5_people'\
+'.csv'
with open(datafilename,'w') as csv_file:
    writer = csv.writer(csv_file, delimiter=',')
    writer.writerow(["person_ID", \
                     "age", \
                     'mutation_ID', \
                     "mutation_type", \
                     "true_freq", \
                     "freq1",\
                     "var_reads1", \
                     "depth1", \
                     "freq2",\
                     "var_reads2", \
                     "depth2"])
    number_alive_at_end=[]
    for person_ID in range(number_of_individuals):
        if person_ID%1==0:
            print(person_ID)
        
        n0=eq_population_size # starting number of cells in first clone
        founding_pop = [  [[(0, 0.0),],n0]   ]
        development_phase = 1
        t = 0.0
        
        population=founding_pop
        population_size=n0

        age_of_blood_draw_1=round(np.random.normal(mean_age, SD_age))
        age_of_blood_draw_2=age_of_blood_draw_1+5
        age_of_blood_draw_3=age_of_blood_draw_1+10
        age_of_blood_draw_4=age_of_blood_draw_1+15

        while t < lifespan:
            if population_size < eq_population_size and development_phase==1:
                t = t+dt
                B0 = 1
                D0 = 0
                population_size=pop_size(population)

                new_population=[]
                for clone_info in population:
                    #first put the ID and clone size in our 'clone histories' dictionary
                    clone_ID=tuple(clone_info[0])
                    clone_size=clone_info[1]
                    if clone_ID not in clone_histories:
                        clone_histories[clone_ID] = {'sizes':[clone_size], 'times':[t]}
                    else:
                        clone_histories[clone_ID]['sizes'].append(clone_size)
                        clone_histories[clone_ID]['times'].append(t)
                    #then update clone sizes/ population list
                    updated_clone=divide(clone_info, dt, B0, D0)
                    (new_clones, new_latest_ID)=mutate(clone_info, dt, u,\
                                                       last_mutation_ID, DFE)
                    last_mutation_ID = new_latest_ID
                    new_population = new_population+new_clones
                    if updated_clone[1]>0:
                        new_population.append(updated_clone)
                population=new_population

            else:
                development_phase=0 #stops the dev phase first time popsize > eq_pop_size 
                
                t = t+dt
                B0 = 0.2
                D0 = 0.2
                population_size=pop_size(population)

                new_population=[]
                for clone_info in population:
                    
                    clone_ID=tuple(clone_info[0])
                    clone_size=clone_info[1]
                    
                    if person_ID<50:
                        #first put the ID and clone size in our 'clone histories' dictionary
                        if t%1<dt:
                            if clone_ID not in clone_histories:
                                clone_histories[clone_ID] = {'sizes':[clone_size], 'times':[t]}
                            else:
                                clone_histories[clone_ID]['sizes'].append(clone_size)
                                clone_histories[clone_ID]['times'].append(t)
                        
                        
                    #then update clone sizes/ population list
                    updated_clone=divide(clone_info, dt, B0, D0)
                    (new_clones, new_latest_ID)=mutate(clone_info, dt, u, last_mutation_ID, DFE)
                    last_mutation_ID = new_latest_ID
                    new_population = new_population+new_clones
                    if updated_clone[1]>0:
                        new_population.append(updated_clone)
                population=new_population

                time_1=int(age_of_blood_draw_1*lam/dt)
                time_2=int(age_of_blood_draw_2*lam/dt)
                time_3=int(age_of_blood_draw_3*lam/dt)
                time_4=int(age_of_blood_draw_4*lam/dt)


                sample_timings=[time_1, time_2, time_3, time_4]

                if int(t/dt) in sample_timings:

                    population_dict[person_ID]=population
    
                    variants=sequence(population, average_depth)
                    for var in variants:                            
                            mID=var['mutation_ID']
                            mutation_type=var['mutation_type']
                            true_freq=var['true_freq']
                            freq1=var['freq1']
                            freq2=var['freq2']
                            reads1=var['reads1']
                            reads2=var['reads2']
                            depth1=var['depth1']
                            depth2=var['depth2']
                            age=int(t/lam)
                    
                            writer.writerow([person_ID, \
                                             age, \
                                             mID, \
                                             mutation_type, \
                                             true_freq, \
                                             freq1,\
                                             reads1, \
                                             depth1, \
                                             freq2,\
                                             reads2, \
                                             depth2])
                        
        number_alive_at_end.append(len(population))
```

```python
# PARAMETERS
eq_population_size = 10**5 #number of stem cells (HSCs)
lifespan = 500 #measured in cell divisions (e.g. 100 = 100 cell divisions)
dt=1.0 # measured in units of the overall cell division rate
u=3*10**(-5) #mutation rate
lam=5 # HSC divisions per year (symmetric and asymmetric)
last_mutation_ID=0
average_depth=3000 #sequencing depth
number_of_individuals=1000
mean_age=50
SD_age=0.01

#FITNESS EFFECTS (per year)
sb=0.01 #beneficial fitness effect
sd=0.00 #deleterious fitness effect
R882 = sb*10 #fitness effect of R882 variant

#PROBABILITY OF EACH TYPE OF VARIANT
Pn=0.7 #probability of neutral variant
PR882 = (1-Pn)/3 #probability of R882 variant
Pnonsyn = ((1-Pn)-PR882) #probability of nonsynonymous variant
Pd=0.00 #probability of deleterious variant

#exponential distribution of fitness effects for nonsynonymous variants
s_max=0.099 #maximum fitness effect for nonsynonymous variant
s_range=np.linspace(sb, s_max, 10) #i.e. nonsynonymous beneficial fitness effects range from sb (=0.01) to s_max (0.099)
d = 0.03
weights=np.array([np.exp(-s/d) for s in s_range])
normalization=np.sum(weights)
norm_weights=(weights/normalization)*Pnonsyn

#DISTRIBUTION OF FITNESS EFFECTS
DFE = [[s, w] for s, w in zip(s_range, norm_weights)]
DFE.append([R882, PR882])
DFE.append([0.0, Pn])
```

```python
#dynamics - watson et al
population_dict={}

timestamp=str(datetime.datetime.now())
datafilename='watson_et_al='+"{0:.2e}".format(average_depth)\
+'_N='+"{0:.2e}".format(eq_population_size)\
+'_U='+"{0:.3e}".format(u)\
+'_s='+str(sb)\
+'_100000people_ages25_50_75_sd0.01'\
+'_expo_dist_s_0.01_0.099'\
+'R882_0.10s'\
+'Pnonsyn_0.4'\
+'.csv'

with open(datafilename,'w') as csv_file:
    writer = csv.writer(csv_file, delimiter=',')
    writer.writerow(["person_ID", \
                     "age", \
                     'mutation_ID', \
                     "mutation_type", \
                     "true_freq", \
                     "freq1",\
                     "var_reads1", \
                     "depth1", \
                     "freq2",\
                     "var_reads2", \
                     "depth2"])
    number_alive_at_end=[]
    for person_ID in range(number_of_individuals):
        print(person_ID)

        n0=eq_population_size # starting number of cells in first clone (e.g. 10^5)
        founding_pop = [  [[(0, 0.0),],n0]   ]
        development_phase = 1
        t = 0.0
        clone_histories={}
        
        population=founding_pop  #[  [[(0, 0.0),],n0]   ]
        population_size=n0  # starts at 10^5

        age_of_blood_draw_1=25  #picks a random age from the normal distribution
        age_of_blood_draw_2=50
        age_of_blood_draw_3=75

        while t < lifespan: # t = 0 at the start
            if population_size < eq_population_size and development_phase==1:
                # when "population_size" first surpasses "eq_population_size" end development phase

                t = t+dt
                B0 = 1
                D0 = 0
                population_size=pop_size(population)

                new_population=[]
                for clone_info in population:
                    #first put the ID and clone size in our 'clone histories' dictionary
                    clone_ID=tuple(clone_info[0])
                    clone_size=clone_info[1]
                    if clone_ID not in clone_histories:
                        clone_histories[clone_ID] = {'sizes':[clone_size], 'times':[t]}
                    else:
                        clone_histories[clone_ID]['sizes'].append(clone_size)
                        clone_histories[clone_ID]['times'].append(t)
                    #then update clone sizes/ population list
                    updated_clone=divide(clone_info, dt, B0, D0)
                    (new_clones, new_latest_ID)=mutate(clone_info, dt, u,\
                                                       last_mutation_ID, DFE)
                    last_mutation_ID = new_latest_ID
                    new_population = new_population+new_clones
                    if updated_clone[1]>0:
                        new_population.append(updated_clone)
                population=new_population

            else:
                development_phase=0 #stops the dev phase first time popsize > eq_pop_size 
                
                t = t+dt
                B0 = 0.2
                D0 = 0.2
                population_size=pop_size(population)

                new_population=[]
                for clone_info in population:
                    
                    clone_ID=tuple(clone_info[0])
                    clone_size=clone_info[1]
                    
                    if person_ID<5:
                        #first put the ID and clone size in our 'clone histories' dictionary
                        if t%1<dt:
                            if clone_ID not in clone_histories:
                                clone_histories[clone_ID] = {'sizes':[clone_size], 'times':[t]}
                            else:
                                clone_histories[clone_ID]['sizes'].append(clone_size)
                                clone_histories[clone_ID]['times'].append(t)
                        
                        
                    #then update clone sizes/ population list
                    updated_clone=divide(clone_info, dt, B0, D0)
                    (new_clones, new_latest_ID)=mutate(clone_info, dt, u, last_mutation_ID, DFE)
                    last_mutation_ID = new_latest_ID
                    new_population = new_population+new_clones
                    if updated_clone[1]>0:
                        new_population.append(updated_clone)
                population=new_population

                time_1=int(age_of_blood_draw_1*lam/dt)
                time_2=int(age_of_blood_draw_2*lam/dt)
                time_3=int(age_of_blood_draw_3*lam/dt)

                sample_timings=[time_1, time_2, time_3]

                if int(t/dt) in sample_timings:

                    population_dict[person_ID]=population
    
                    variants=sequence(population, average_depth)
        
                    for var in variants:                            
                            mID=var['mutation_ID']
                            mutation_type=var['mutation_type']
                            true_freq=var['true_freq']
                            freq1=var['freq1']
                            freq2=var['freq2']
                            reads1=var['reads1']
                            reads2=var['reads2']
                            depth1=var['depth1']
                            depth2=var['depth2']
                            age=int(t/lam)
                    
                            writer.writerow([person_ID, \
                                             age, \
                                             mID, \
                                             mutation_type, \
                                             true_freq, \
                                             freq1,\
                                             reads1, \
                                             depth1, \
                                             freq2,\
                                             reads2, \
                                             depth2])
                        
        number_alive_at_end.append(len(population))
        
        if person_ID<2:
            #Plotting the clonal dynamics/ clone sizes over time

            plt.close('all')
            f, (ax) = plt.subplots(1, 1, sharey=False,figsize=(10,7))
            font = {'weight' : 'normal','size'   : 16}
            font1 = {'weight' : 'normal','size'   : 16}
            matplotlib.rc('font', **font)

            ax.plot([t for t in range(-2, lifespan)], [0.1*eq_population_size for t in range(-2, lifespan)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([t for t in range(-2, lifespan)], [0.01*eq_population_size for t in range(-2, lifespan)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([t for t in range(-2, lifespan)], [0.001*eq_population_size for t in range(-2, lifespan)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([t for t in range(-2, lifespan)], [0.0001*eq_population_size for t in range(-2, lifespan)], lw=1, linestyle = 'dashed', color='k')

            ax.plot([5*60 for t in range(100)], [10**(t/16) for t in range(100)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([5*65 for t in range(100)], [10**(t/16) for t in range(100)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([5*70 for t in range(100)], [10**(t/16) for t in range(100)], lw=1, linestyle = 'dashed', color='k')
            ax.plot([5*75 for t in range(100)], [10**(t/16) for t in range(100)], lw=1, linestyle = 'dashed', color='k')

            for clone_ID, entry in clone_histories.items():
                fitness=sum(i[1] for i in clone_ID)
                sizes = entry["sizes"]
                times = entry["times"]
                
                if sb <= fitness <= 10*sb: #missense and nonsense
                    col=nonsyncolor
                    linew=1.2

                elif 10.1*sb < fitness: #don't show studies with more than one mutation
                    col=c8
                    linew=0.0 
                elif fitness ==sb*10: #R882
                    col=R882color
                    linew=1.2
                elif fitness==0.0:#neutral variants
                    col = neutralcolor
                    linew = 1.2
                if clone_ID!= ((0, 0.0),):
                    ax.plot(times, sizes, color=col, lw=linew)

            ax.set_xlabel('time (cell divisions)')
            ax.set_yscale('log')
            ax.set_ylabel('clone size (cells)')
            ax.set_title('Clone Size over Time: Beneficial fitness effect ='+ str(sb), fontdict=font1)
            ax.set_xlim(0, lifespan)
            ax.set_ylim(1, 10**5.2);
            ax.text(lifespan+2, 0.08*eq_population_size, r'$10\%$', fontdict=font1)
            ax.text(lifespan+2, 0.008*eq_population_size, r'$1\%$', fontdict=font1)
            ax.text(lifespan+2, 0.0008*eq_population_size, r'$0.1\%$', fontdict=font1)
            ax.text(lifespan+2, 0.00008*eq_population_size, r'$0.01\%$', fontdict=font1)
            plt.savefig('trajectories_'+str(person_ID)+'.pdf',format = 'pdf', transparent=True, bbox_inches='tight')
            plt.show()
```

# Plotting the distribution of VAFs at age 25, 50 and 75 years old

```python
def weighted_rho(x, t, d, N, c): #function for exponential distribution of fitness effects (for range of s in nonsynonymous variants)
    s_max=0.10
    s_range=np.linspace(0.01, s_max, 10) #fitness effects for nonsynonymous range from 0.01 to s_max (0.1)
    weights=np.array([np.exp(-s/d) for s in s_range]) #exponential distribution of fitness effects (in range 0.01 to s_max)
    normalization=np.sum(weights)
    norm_weights=weights/normalization
    rho_components=0.0
    for weight, s in zip(norm_weights, s_range):
        phi=c*((np.exp(s*t*lam*D0)-1)/(2*N*s))
        rho_components=rho_components+weight*np.exp(-np.exp(x)/(phi*(1-2*np.exp(x))))
        
    rho_x = np.log((N/(1-2*np.exp(x)))*rho_components)

    return rho_x
```
