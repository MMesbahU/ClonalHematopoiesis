# Plotting the variants across DNMT3A

```python
#Data - create a dictionary that contains info on all the variants as well as separate dictionaries for each study
import csv
filename="DNMT3A_variant_positions.csv" #csv file containing VAFs for each SNV (in each study) with its associated AA position

with open(filename, 'r') as csvfile:
    read_reader = csv.reader(csvfile)  #csv.reader returns a reader object which will iterate over lines in the csvfile
    row_count=0
    Jaiswal_variant_positions = {}
    Genovese_variant_positions = {}
    Zink_variant_positions = {}
    McKerrel_variant_positions = {}
    Acuna_variant_positions = {}
    Coombs_variant_positions = {}
    Young_variant_positions = {}
    Desai_variant_positions = {}
    all_variant_positions = {}
    
    for row in read_reader:
        position = row[2] #amino acid position in DNMT3A
        VAF = row[3]
        study = row[0] #study name
        variant_type = row[4] #missense, nonsense or synonymous
        if study == 'Jaiswal 2014':
            Jaiswal_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Genovese 2014':
            Genovese_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Zink 2017':
            Zink_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'McKerrel 2015':
            McKerrel_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Acuna-Hidalgo 2017':
            Acuna_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Coombs 2017':
            Coombs_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Young 2016':
            Young_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Young 2019':
            Young_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)
        if study == 'Desai 2018':
            Desai_variant_positions[row_count]=(study, position, VAF, variant_type)
            all_variant_positions[row_count]=(study, position, VAF, variant_type)

        row_count=row_count+1
```
